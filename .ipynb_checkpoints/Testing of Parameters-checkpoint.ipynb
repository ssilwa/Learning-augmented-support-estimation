## Base experiment code

```python

def base_tester(data, bases, num_iters, percent, print_int = False, L = 0):

    # load data
    dataset, data_type = data
    
    print("dataset: " + dataset + ", type:" + data_type)

    
    # actual data
    actual_file = np.load('data/actual/' + dataset + '/' + dataset + '_actual_counts_' +  data_type + '.npz')
    actual_counts = actual_file['arr_0']
    actual_prob = actual_counts/actual_counts.sum()
    
    # set parameters
    n = actual_counts.sum()
    act_supp = len(actual_counts)
    print("n: {}".format(n))
    print("actual support: {}".format(act_supp))
    print("percent: {}".format(percent))
    
    sample_size = int(np.random.poisson(n*percent))
#     sample_size = int(n*percent)
    print("sample size: {}".format(sample_size))
    WY_estimator = WYSupport(pmin=1./n)
    
    
    
    # predicted data
    predicted_file = np.load('data/predictions/' + dataset + '/' + dataset + '_oracle_' + data_type + '.npz')
    try:
        predicted_counts = predicted_file['test_output'][:,0]
    except:
        predicted_counts = predicted_file['arr_0']
        
    # ip predictions are originally logs
    if dataset == "ip":
        predicted_counts = np.exp(predicted_counts)
    
    
    predicted_counts[ predicted_counts < 1.0] = 1.0
    predicted_prob = predicted_counts/n
    
    oracle_holder = np.zeros((num_iters, len(bases)))
    naive_holder = np.zeros((num_iters, len(bases)))
    WY_holder = np.zeros(num_iters)
    failure_holder = np.zeros((num_iters, len(bases)))
    interval_holder = np.zeros((num_iters, len(bases)))
    
    for i in range(num_iters):
        print(i)
        samples = np.random.choice(act_supp, size=sample_size, p=actual_prob)
        histogram = samples_to_histogram(samples)
        
        pred_array_oracle = []
        pred_array_naive = []
        failure_array = []
        interval_array = []
        
        for base in bases:
            oracle_estimate_pred, naive_oracle_estimate_pred, failure, num_intervals = callOracleEstimator(histogram, predicted_counts, base, n, sample_size, L)
            pred_array_oracle.append(oracle_estimate_pred)
            pred_array_naive.append(naive_oracle_estimate_pred)
            failure_array.append(failure)
            interval_array.append(num_intervals)
            
            if print_int:
                print(base, num_intervals, failure)
        
        oracle_holder[i,:] = pred_array_oracle
        naive_holder[i,:] = pred_array_naive
        failure_holder[i,:] = failure_array
        interval_holder[i,:] = interval_array
        
        
        fin = sample_to_fin(samples)
        WY_estimate = WY_estimator.estimate(fin)
        WY_holder[i] = WY_estimate
    
    # take median
    oracle_holder = np.abs(1-oracle_holder/act_supp)
    naive_holder = np.abs(1-naive_holder/act_supp)
    oracle_output = np.median(oracle_holder, axis = 0)
    
    naive_output = np.median(naive_holder, axis = 0)
    std_oracle_output = np.std(oracle_holder, axis = 0)
    WY_holder = np.abs(1-WY_holder/act_supp)
    wy_output = np.median(WY_holder)
    wy_std = np.std(WY_holder)
    
    failure_output = np.median(failure_holder, axis = 0)
    interval_output = np.median(interval_holder, axis = 0)
            
    return oracle_output, std_oracle_output, wy_output, failure_output, interval_output  #naive_output, std_oracle_output, wy_output, wy_std 

def base_tester_zipf(bases, actual_prob, predicted_counts, sample_size, n, L, n_iter=5, print_int = False):
    output = np.zeros((n_iter, len(bases)))
    output_failure = np.zeros((n_iter, len(bases)))
    output_interval = np.zeros((n_iter, len(bases)))
    
    for i in range(n_iter):
        print(i)
        samples = np.random.choice(act_supp, size=sample_size, p=actual_prob)
        histogram = samples_to_histogram(samples)
        
        curr_result_oracle = []
        #curr_result_naive = []
        wy_estimates = []
        failure_oracle = []
        interval_oracle = []
        
        curr_wy_estimator = WYSupport(pmin)
        curr_wy_estimate = curr_wy_estimator.estimate(sample_to_fin(samples))
        
        for base in bases:
            oracle_estimate_pred, naive_oracle_estimate_pred, failure, num_intervals = callOracleEstimator(histogram, predicted_counts, base, n, sample_size, L)
            curr_result_oracle.append(oracle_estimate_pred)
            failure_oracle.append(failure)
            interval_oracle.append(num_intervals)
            
            
            if print_int:
                print(num_intervals, failure)
            
        output[i,:] = curr_result_oracle
        output_failure[i,:] = failure_oracle
        output_interval[i,:] = interval_oracle
        wy_estimates.append(curr_wy_estimate)
    output = np.abs(1-output/act_supp)
    output_oracle = np.median(output, axis=0)
    output_std = np.std(output, axis = 0)
    output_failure = np.median(output_failure, axis = 0)
    output_interval = np.median(output_interval, axis = 0)
    wy_estimates = [np.abs(1-k/act_supp) for k in wy_estimates]
        
    return output_oracle, output_std, np.median(wy_estimates), output_failure, output_interval
```

# Choosing optimal base at each trial

```python
def base_chooser(bases, failure, error):
    
    best_indx = 0
    for indx, fail_perct in enumerate(failure):
        if sum(failure[indx:]) == 0.0:
            best_index = indx
            break
    
    max_index = len(bases)
    # old method ->>> best_index2 = int(min(best_index*1.75, len(bases)-1))
    # best_index2 = int((max_index - best_index)*.5) + best_index

    if best_index*2 >= max_index-1:
        best_index2 = int(min(best_index*1.5, max_index-1))
    else:
        best_index2 = int(best_index*2)

    return (bases[best_index2], error[best_index2])
    





def base_tester(data, bases, num_iters, percent, print_int = False, L = 0):

    # load data
    dataset, data_type = data
    
    print("dataset: " + dataset + ", type:" + data_type)

    
    # actual data
    actual_file = np.load('data/actual/' + dataset + '/' + dataset + '_actual_counts_' +  data_type + '.npz')
    actual_counts = actual_file['arr_0']
    actual_prob = actual_counts/actual_counts.sum()
    
    # set parameters
    n = actual_counts.sum()
    act_supp = len(actual_counts)
    print("n: {}".format(n))
    print("actual support: {}".format(act_supp))
    print("percent: {}".format(percent))
    sample_size = int(n*percent)
    print("sample size: {}".format(sample_size))
    WY_estimator = WYSupport(pmin=1./n)
    
    
    
    # predicted data
    predicted_file = np.load('data/predictions/' + dataset + '/' + dataset + '_oracle_' + data_type + '.npz')
    try:
        predicted_counts = predicted_file['test_output'][:,0]
    except:
        predicted_counts = predicted_file['arr_0']
        
    print(predicted_counts.shape)
        
    # ip predictions are originally logs
    if dataset == "ip":
        predicted_counts = np.exp(predicted_counts)
    
    
    predicted_counts[ predicted_counts < 1.0] = 1.0
    predicted_prob = predicted_counts/n
    
    oracle_holder = np.zeros((num_iters, len(bases)))
    naive_holder = np.zeros((num_iters, len(bases)))
    WY_holder = np.zeros(num_iters)
    failure_holder = np.zeros((num_iters, len(bases)))
    interval_holder = np.zeros((num_iters, len(bases)))
    optimal_error_holder = np.zeros(num_iters)
    optimal_base_holder = np.zeros(num_iters)
    
    for i in range(num_iters):
        print(i)
        samples = np.random.choice(act_supp, size=sample_size, p=actual_prob)
        histogram = samples_to_histogram(samples)
        
        pred_array_oracle = []
        pred_array_naive = []
        failure_array = []
        interval_array = []
        
        for base in bases:
            oracle_estimate_pred, naive_oracle_estimate_pred, failure, num_intervals = callOracleEstimator(histogram, predicted_counts, base, n, sample_size, L)
            pred_array_oracle.append(oracle_estimate_pred)
            pred_array_naive.append(naive_oracle_estimate_pred)
            failure_array.append(failure)
            interval_array.append(num_intervals)
            
            if print_int:
                print(base, num_intervals, failure)
        
        oracle_holder[i,:] = pred_array_oracle
        naive_holder[i,:] = pred_array_naive
        failure_holder[i,:] = failure_array
        interval_holder[i,:] = interval_array
        curr_error =  [abs(1-prediction/act_supp) for prediction in pred_array_oracle] 
        
        
        best_base, base_error = base_chooser(bases, failure_array, curr_error)
        optimal_error_holder[i] = base_error
        optimal_base_holder[i] = best_base
        
        
        
        fin = sample_to_fin(samples)
        WY_estimate = WY_estimator.estimate(fin)
        WY_holder[i] = WY_estimate
        
#         print(best_base, base_error, abs(1-WY_estimate/act_supp))

    
    # take median
    oracle_holder = np.abs(1-oracle_holder/act_supp)
    naive_holder = np.abs(1-naive_holder/act_supp)
    oracle_output = np.median(oracle_holder, axis = 0)
    
    naive_output = np.median(naive_holder, axis = 0)
    std_oracle_output = np.std(oracle_holder, axis = 0)
    WY_holder = np.abs(1-WY_holder/act_supp)
    wy_output = np.median(WY_holder)
    wy_std = np.std(WY_holder)
    
    failure_output = np.median(failure_holder, axis = 0)
    interval_output = np.median(interval_holder, axis = 0)
    
    error_output = np.median(optimal_error_holder)
    error_std = np.std(optimal_error_holder)
    base_output = np.median(optimal_base_holder)
    
            
    return oracle_output, std_oracle_output, wy_output, wy_std, failure_output, interval_output, naive_output, error_output, error_std, base_output

def base_tester_zipf(bases, actual_prob, predicted_counts, sample_size, n, L, n_iter=5, print_int = False):
    output = np.zeros((n_iter, len(bases)))
    output_failure = np.zeros((n_iter, len(bases)))
    output_interval = np.zeros((n_iter, len(bases)))
    optimal_error_holder = np.zeros(n_iter)
    optimal_base_holder = np.zeros(n_iter)
    
    for i in range(n_iter):
        samples = np.random.choice(act_supp, size=sample_size, p=actual_prob)
        histogram = samples_to_histogram(samples)
        
        curr_result_oracle = []
        #curr_result_naive = []
        wy_estimates = []
        failure_oracle = []
        interval_oracle = []
        
        curr_wy_estimator = WYSupport(pmin)
        curr_wy_estimate = curr_wy_estimator.estimate(sample_to_fin(samples))
        
        for base in bases:
            oracle_estimate_pred, naive_oracle_estimate_pred, failure, num_intervals = callOracleEstimator(histogram, predicted_counts, base, n, sample_size, L)
            curr_result_oracle.append(oracle_estimate_pred)
            failure_oracle.append(failure)
            interval_oracle.append(num_intervals)
            
            
            if print_int:
                print(num_intervals, failure)
            
        output[i,:] = curr_result_oracle
        output_failure[i,:] = failure_oracle
        output_interval[i,:] = interval_oracle
        wy_estimates.append(curr_wy_estimate)
        curr_error =  [abs(1-prediction/act_supp) for prediction in curr_result_oracle] 

                
        best_base, base_error = base_chooser(bases, failure_oracle, curr_error)
        optimal_error_holder[i] = base_error
        optimal_base_holder[i] = best_base
        
    output = np.abs(1-output/act_supp)
    output_oracle = np.median(output, axis=0)
    output_std = np.std(output, axis = 0)
    output_failure = np.median(output_failure, axis = 0)
    output_interval = np.median(output_interval, axis = 0)
    wy_estimates = [np.abs(1-k/act_supp) for k in wy_estimates]
    
    error_output = np.median(optimal_error_holder)
    error_std = np.std(optimal_error_holder)
    base_output = np.median(optimal_base_holder)
    
    return output_oracle, output_std, np.median(wy_estimates), np.std(wy_estimates), output_failure, output_interval, optimal_error_holder, error_std, base_output
```

```python
def generate_zip_prob(n, alpha=1.0):
    output = np.arange(1.0, n+1.0)
    output = output**(-alpha)
    const = output.sum()
    output = (1./const)*output
    return output

act_supp = 100000
actual_prob = generate_zip_prob(act_supp, alpha=.5)
pmin = actual_prob[-1]
n = int(1./pmin)

samples_for_pred = np.random.choice(act_supp, size=int(n*.1), p=actual_prob)
samp_hist = {}
for sample in samples_for_pred:
    if sample not in samp_hist:
        samp_hist[sample] = 0
    samp_hist[sample] += 1

predicted_counts = []
for i in range(act_supp):
    if i in samp_hist:
        predicted_counts.append(samp_hist[i])
    else:
        predicted_counts.append(1)
        
```

```python
def base_tester(data, bases, num_iters, percent, print_int = False, L = 0):

    # load data
    dataset, data_type = data
    
    print("dataset: " + dataset + ", type:" + data_type)

    
    # actual data
    actual_file = np.load('data/actual/' + dataset + '/' + dataset + '_actual_counts_' +  data_type + '.npz')
    actual_counts = actual_file['arr_0']
    actual_counts = actual_counts[:78500]
    actual_prob = actual_counts/actual_counts.sum()
    
    # set parameters
    n = actual_counts.sum()
    act_supp = len(actual_counts)
    print("n: {}".format(n))
    print("actual support: {}".format(act_supp))
    print("percent: {}".format(percent))
    sample_size = int(n*percent)
    print("sample size: {}".format(sample_size))
    WY_estimator = WYSupport(pmin=1./n)
    
    
    
    # predicted data
    predicted_file = np.load('data/predictions/' + dataset + '/' + dataset + '_oracle_' + data_type + '.npz')
    try:
        predicted_counts = predicted_file['test_output'][:,0]
    except:
        predicted_counts = predicted_file['arr_0']
        
    print(predicted_counts.shape)
    
    predicted_counts = predicted_counts[:78500]
        
    # ip predictions are originally logs
    if dataset == "ip":
        predicted_counts = np.exp(predicted_counts)
    
    
    predicted_counts[ predicted_counts < 1.0] = 1.0
    predicted_prob = predicted_counts/n
    
    wy_est_holder = np.zeros((num_iters, len(bases)))
    naive_holder = np.zeros((num_iters, len(bases)))
    distinct_holder = np.zeros((num_iters, len(bases)))
    WY_holder = np.zeros(num_iters)
    failure_holder = np.zeros((num_iters, len(bases)))
    interval_holder = np.zeros((num_iters, len(bases)))
    
    for i in range(num_iters):
        print(i)
        samples = np.random.choice(act_supp, size=sample_size, p=actual_prob)
        histogram = samples_to_histogram(samples)
        
        wy_array = []
        naive_array = []
        distinct_array = []
        failure_array = []
        interval_array = []
        
        for base in bases:
            
            wy_estimate_pred, naive_estimate_pred, distinct_estimate_pred, failure, num_intervals = callOracleEstimator(histogram, predicted_counts, base, n, sample_size, L)
            wy_array.append(wy_estimate_pred)
            naive_array.append(naive_estimate_pred)
            distinct_array.append(distinct_estimate_pred)
            failure_array.append(failure)
            interval_array.append(num_intervals)
            
            if print_int:
                print(base, num_intervals, failure)
        
        wy_est_holder[i,:] = wy_array
        naive_holder[i,:] = naive_array
        distinct_holder[i,:] = distinct_array
        failure_holder[i,:] = failure_array
        interval_holder[i,:] = interval_array        
        
        fin = sample_to_fin(samples)
        WY_estimate = WY_estimator.estimate(fin)
        WY_holder[i] = WY_estimate
        
    
    # take median
    wy_est_holder = np.abs(1-wy_est_holder/act_supp)
    naive_holder = np.abs(1-naive_holder/act_supp)
    distinct_holder = np.abs(1-distinct_holder/act_supp)
    
    wy_est_output = np.median(wy_est_holder, axis = 0)
    naive_output = np.median(naive_holder, axis = 0)
    distinct_output = np.median(distinct_holder, axis = 0)
    
    
    wy_est_output_std = np.std(wy_est_holder, axis = 0)
    naive_output_std = np.std(naive_holder, axis = 0)
    distinct_output_std = np.std(distinct_holder, axis = 0)
    
    WY_holder = np.abs(1-WY_holder/act_supp)
    wy_output = np.median(WY_holder)
    wy_std = np.std(WY_holder)
    
    failure_output = np.median(failure_holder, axis = 0)
    interval_output = np.median(interval_holder, axis = 0)
            
    return wy_est_output, wy_est_output_std, naive_output, naive_output_std, distinct_output, distinct_output_std, failure_output, interval_output, wy_output


def OracleEstimator(histograms, intervals, L, n, n_samples, base):
    """
    Returns sum of supports for each interval
    
    Args:
    histograms: list of histograms, one for each interval
    intervals: list of intervals where each interval is of the form [left_end, right_end]
    L: degree of polynomial
    """
    
    num_intervals = len(intervals)
    num_failures = 0
    supp_estimate_WY = 0.0
    supp_estimate_naive = 0.0
    supp_estimate_distinct = 0.0

    # loop over histogram and intervals and add up support 
    for i, hist in enumerate(histograms):
        use_naive = True
        curr_interval = intervals[i]
        curr_estimate = estimator(L, hist, curr_interval)
        
        if np.isnan(curr_estimate) or curr_estimate < 0 or curr_estimate > 1./curr_interval[0]:
            num_failures += 1
            supp_estimate_WY += curr_estimate
            
            midpoint = (intervals[i][0]+intervals[i][1])*1./2
            naive_estimate = sum(hist)*1./(midpoint * n_samples)
            distinct_estimate = len(hist)
            
            supp_estimate_naive += naive_estimate
            supp_estimate_distinct += distinct_estimate
        else:
            supp_estimate_WY += curr_estimate
            supp_estimate_naive += curr_estimate
            supp_estimate_distinct += curr_estimate
            

    return [supp_estimate_WY, supp_estimate_naive, supp_estimate_distinct, num_failures/num_intervals, num_intervals]

def callOracleEstimator(histogram, oracle_counts, base, n, n_samples, L):
    if L == 0:
        L = int(floor(0.45*log(n)))

        
    hist_dic = {}
    outlier = []
    
    max_val = 0.5*n*log(n)/n_samples
    base_log = log(base)
    log_vals = np.floor(np.log(oracle_counts)/base_log)
    
    for sample, value in histogram.items():
        
        if oracle_counts[sample] <= max_val:
            p = log_vals[sample] 
            if p not in hist_dic:
                hist_dic[p] = []
            hist_dic[p].append(value)
        else:
            outlier.append(value)
    
    
    max_slots = int(max(hist_dic.keys()))+1
    intervals = [[] for k in range(max_slots)]
    histograms = [[] for k in range(max_slots)]
    
    for p,h in hist_dic.items():
        eff_p = base**p
        intervals[int(p)] = [eff_p*1./n, eff_p*base*1./n]
        #intervals[int(p)] = [eff_p*1./n, min(0.5*n*log(n)/n_samples, eff_p*base)*1./n]
        histograms[int(p)] += h
        
            
    
    max_p = int(np.floor(np.log(0.5*n*log(n)/n_samples)/np.log(base)))

    if max_p > max_slots-1:
        eff_p = base**max_p
        intervals.append([eff_p*1./n, eff_p*base*1./n])
        #intervals.append([eff_p*1./n, min(0.5*n*log(n)/n_samples, eff_p*base)*1./n])
        histograms.append(outlier)
    else:
        histograms[max_p] += outlier
    
    histograms = [hists for hists in histograms if len(hists) > 0]
    intervals = [intvs for intvs in intervals if len(intvs) > 0]

            
    output = OracleEstimator(histograms, intervals, L, n, n_samples, base)
    return output
```

```python
def base_tester_zipf(bases, actual_prob, predicted_counts, sample_size, n, L, n_iter=5, print_int = False):
    output = np.zeros((n_iter, len(bases)))
    output_failure = np.zeros((n_iter, len(bases)))
    output_interval = np.zeros((n_iter, len(bases)))

    
    for i in range(n_iter):
        print(i)
        samples = np.random.choice(act_supp, size=sample_size, p=actual_prob)
        histogram = samples_to_histogram(samples)
        
        curr_result_oracle = []
        wy_estimates = []
        failure_oracle = []
        interval_oracle = []
        
        curr_wy_estimator = WYSupport(pmin)
        curr_wy_estimate = curr_wy_estimator.estimate(sample_to_fin(samples))
        
        for base in bases:
                        
            
            junk1, junk2, oracle_estimate_pred, failure, num_intervals = callOracleEstimator(histogram, predicted_counts, base, n, sample_size, L)
            
            curr_result_oracle.append(oracle_estimate_pred)
            failure_oracle.append(failure)
            interval_oracle.append(num_intervals)
            
            
            if print_int:
                print(num_intervals, failure)
            
        output[i,:] = curr_result_oracle
        output_failure[i,:] = failure_oracle
        output_interval[i,:] = interval_oracle
        wy_estimates.append(curr_wy_estimate)

                
        
    output = np.abs(1-output/act_supp)
    output_oracle = np.median(output, axis=0)
    output_std = np.std(output, axis = 0)
    output_failure = np.median(output_failure, axis = 0)
    output_interval = np.median(output_interval, axis = 0)
    wy_estimates = [np.abs(1-k/act_supp) for k in wy_estimates]
    
    return output_oracle, output_std, np.median(wy_estimates), np.std(wy_estimates), output_failure, output_interval
```

```python
def generate_zip_prob(n, alpha=1.0):
    output = np.arange(1.0, n+1.0)
    output = output**(-alpha)
    const = output.sum()
    output = (1./const)*output
    return output

act_supp = 100000
actual_prob = generate_zip_prob(act_supp, alpha=.5)
pmin = actual_prob[-1]
n = int(1./pmin)

samples_for_pred = np.random.choice(act_supp, size=int(n*.1), p=actual_prob)
samp_hist = {}
for sample in samples_for_pred:
    if sample not in samp_hist:
        samp_hist[sample] = 0
    samp_hist[sample] += 1

predicted_counts = []
for i in range(act_supp):
    if i in samp_hist:
        predicted_counts.append(samp_hist[i])
    else:
        predicted_counts.append(1)
        
```

# Comparision to CR

```python
def cr_alg(samples, predicted_prob, m):
    output = (predicted_prob[samples])**(-1.0)
    return output.sum()/m
    

def cr_tester(data, num_iters, percent):

    # load data
    dataset, data_type = data
    
    print("dataset: " + dataset + ", type:" + data_type)

    
    # actual data
    actual_file = np.load('data/actual/' + dataset + '/' + dataset + '_actual_counts_' +  data_type + '.npz')
    actual_counts = actual_file['arr_0']
    actual_counts = actual_counts[:78500]
    actual_prob = actual_counts/actual_counts.sum()
    
    # set parameters
    n = actual_counts.sum()
    act_supp = len(actual_counts)
    print("n: {}".format(n))
    print("actual support: {}".format(act_supp))
    print("percent: {}".format(percent))
    sample_size = int(n*percent)
    print("sample size: {}".format(sample_size))
    
    
    
    # predicted data
    predicted_file = np.load('data/predictions/' + dataset + '/' + dataset + '_oracle_' + data_type + '.npz')
    try:
        predicted_counts = predicted_file['test_output'][:,0]
    except:
        predicted_counts = predicted_file['arr_0']
        
    print(predicted_counts.shape)
    predicted_counts = predicted_counts[:78500]
        
    # ip predictions are originally logs
    if dataset == "ip":
        predicted_counts = np.exp(predicted_counts)
    
    
    predicted_counts[ predicted_counts < 1.0] = 1.0
    predicted_prob = predicted_counts/n
    
    output = np.zeros(num_iters)
    
    for i in range(num_iters):
        samples = np.random.choice(act_supp, size=sample_size, p=actual_prob)
        output[i] = cr_alg(samples, predicted_prob, sample_size)
    output = np.abs(1-output/act_supp)
        
    return np.median(output), np.std(output)
      
```

```python
# Testing uniform
def generate_zip_prob(n, alpha=1.0):
    output = np.arange(1.0, n+1.0)
    output = output**(-alpha)
    const = output.sum()
    output = (1./const)*output
    return output

act_supp = 100000
actual_prob = generate_zip_prob(act_supp, alpha=.5)
pmin = actual_prob[-1]
n = int(1./pmin)

samples_for_pred = np.random.choice(act_supp, size=int(n*.1), p=actual_prob)
samp_hist = {}
for sample in samples_for_pred:
    if sample not in samp_hist:
        samp_hist[sample] = 0
    samp_hist[sample] += 1

predicted_counts = []
for i in range(act_supp):
    if i in samp_hist:
        predicted_counts.append(samp_hist[i])
    else:
        predicted_counts.append(1)
        
        
```
